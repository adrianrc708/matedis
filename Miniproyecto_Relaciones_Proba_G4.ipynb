```python
import numpy as np
import pandas as pd
from scipy.stats import norm
import matplotlib.pyplot as plt

# Simula la evolución de los precios de los activos
def simular_precio(inicial, volatilidad, dias):
    return inicial * np.cumprod(1 + np.random.normal(0, volatilidad, dias))

# Calcula la probabilidad de pérdida
def calcular_probabilidad_perdida(rendimiento, umbral):
    return norm.cdf(umbral, loc=rendimiento.mean(), scale=rendimiento.std())

# Ejemplo de uso
inicial = 1000
volatilidad = 0.05
dias = 252
umbral = 0.1

precios = simular_precio(inicial, volatilidad, dias)
rendimiento = (precios[1:] - precios[:-1]) / precios[:-1]
probabilidad_perdida = calcular_probabilidad_perdida(rendimiento, umbral)

print(f"Probabilidad de pérdida: {probabilidad_perdida:.2%}")
plt.plot(precios)
plt.show()
```
